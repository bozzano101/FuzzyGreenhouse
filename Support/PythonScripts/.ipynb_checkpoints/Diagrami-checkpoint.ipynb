```python
## TROUGAONA FUNKCIJA

import matplotlib.pyplot as plt
plt.plot([0,5,10,15,20],[0,0,1,0,0])
plt.ylabel(r'$\mu_A(x)$')
plt.tick_params(axis='x',which='both',bottom=False, labelbottom=False)
plt.show()
```

```python
## SIGMOIDNA FUNKCIJA

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-5,5,100)
y = 1 / (1 + math.e ** -x)

plt.plot(x,y)
plt.ylabel(r'$\mu_A(x)$')
plt.tick_params(axis='x',which='both',bottom=False, labelbottom=False)
plt.show()

```

```python
## GAUSOVA FUNKCIJA

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-5,5,100)
y = math.e ** ( -(x**2) )

plt.plot(x,y)
plt.ylabel(r'$\mu_A(x)$')
plt.tick_params(axis='x',which='both',bottom=False, labelbottom=False)
plt.show()
```
